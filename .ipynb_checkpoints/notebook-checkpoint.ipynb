```python
import pandas as pd

df = pd.read_csv("sp 500.csv")

# Date to datetime + sort
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# Columns that should be numeric
num_cols = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]

for c in num_cols:
    # convert to string -> remove commas -> convert to number
    df[c] = (
        df[c].astype(str)
             .str.replace(",", "", regex=False)
             .str.strip()
    )
    df[c] = pd.to_numeric(df[c], errors="coerce")

# drop rows that became NaN after conversion
df = df.dropna(subset=["Close"])

df.dtypes
```

```python
df[["Date","Close"]].head()
```

```python
import numpy as np

# Moving averages
df["SMA3"] = df["Close"].rolling(3).mean()
df["SMA6"] = df["Close"].rolling(6).mean()

# RSI(6) for monthly data
delta = df["Close"].diff()
gain = delta.clip(lower=0).rolling(6).mean()
loss = (-delta.clip(upper=0)).rolling(6).mean()
rs = gain / (loss + 1e-9)
df["RSI6"] = 100 - (100 / (1 + rs))

df = df.dropna()
df[["Date","Close","SMA3","SMA6","RSI6"]].head()
```

```python
df["TomorrowClose"] = df["Close"].shift(-1)
df["Target"] = (df["TomorrowClose"] > df["Close"]).astype(int)
df = df.dropna()
df[["Date","Close","TomorrowClose","Target"]].tail()
```

```python
from sklearn.preprocessing import StandardScaler

features = df[["Close","SMA3","SMA6","RSI6"]].copy()
target = df["Target"].values

scaler = StandardScaler()
scaled = scaler.fit_transform(features)
```

```python
from sklearn.preprocessing import StandardScaler

features = df[["Close","SMA3","SMA6","RSI6"]].copy()
target = df["Target"].values

scaler = StandardScaler()
scaled = scaler.fit_transform(features)
```

```python
import numpy as np

LOOKBACK = 6
X, y = [], []

for i in range(LOOKBACK, len(df)):
    X.append(scaled[i-LOOKBACK:i, :])
    y.append(target[i])

X = np.array(X)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
split = int(len(X) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print("Train:", X_train.shape, y_train.shape)
print("Test :", X_test.shape, y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential([
    LSTM(32, return_sequences=True, input_shape=(LOOKBACK, X.shape[2])),
    Dropout(0.2),
    LSTM(16),
    Dropout(0.2),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
early = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=8,
    validation_split=0.1,
    callbacks=[early],
    verbose=1
)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

pred_prob = model.predict(X_test).ravel()
pred = (pred_prob > 0.5).astype(int)

print("Accuracy:", accuracy_score(y_test, pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, pred))
print(classification_report(y_test, pred, digits=4))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Loss Curve")
plt.show()
```

```python
test_index = df.index[LOOKBACK + split : LOOKBACK + len(X)]  # aligns to X_test
test_dates = df.loc[test_index, "Date"]
test_close = df.loc[test_index, "Close"]

plt.figure(figsize=(12,5))
plt.plot(test_dates, test_close, label="Close")

up_dates = test_dates[pred == 1]
down_dates = test_dates[pred == 0]

plt.scatter(up_dates, test_close[pred == 1], marker="^", label="Pred UP")
plt.scatter(down_dates, test_close[pred == 0], marker="v", label="Pred DOWN")

plt.title("S&P 500 Close with Predicted Trend (Test)")
plt.legend()
plt.show()
```
